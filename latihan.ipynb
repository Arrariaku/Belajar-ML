```python
import requests
import json

ds_list_response = requests.get("http://data.bandung.go.id/api/3/action/package_list")
ds_list_result = json.loads(ds_list_response.text)
len(ds_list_result["result"])
ds_list_result
```

```python
search_response = requests.get("http://data.bandung.go.id/api/3/action/package_search", params=[('q', 'sekolah dasar')])
search_result = json.loads(search_response.text)
print("Result counts: ", search_result["result"]["count"])
print(len(search_result["result"]["results"]))
search_result
```

```python
sd_datasets = [ds_name for ds_name in ds_list_result["result"] if "sekolah" in ds_name and "dasar" in ds_name]
sd_datasets
```

```python
ds_name = "jumlah-guru-sd-negeri-di-kota-bandung-berdasarkan-sekolah"
ds_response = requests.get("http://data.bandung.go.id/api/3/action/package_show", params = [("id", ds_name)])
ds_info = json.loads(ds_response.text)
len(ds_info['result']['resources'])
ds_info
```

```python
ds_urls = [d["url"] for d in ds_info["result"]["resources"]]
ds_urls
```

```python
from tqdm import tqdm
for url in ds_urls:
    fname = url[1 + url.rfind("/"):]
    with open(fname, "wb") as handle:
        resp = requests.get(url, stream = True)
        for data in tqdm(resp.iter_content()):
            handle.write(data)
        print(fname, "Saved.")
```
